### 22. 
**Create a $3\ \times\ 4$ random uniform matrix, then normalize the values to range [0-1]**<pre> (2 lines ★★☆) 

```python
np.random.seed(25)
P = np.around(np.random.uniform(0.0,1.0,(3,2)),decimals=2)
print('before')
print(P,'\n')

Pmax, Pmin = P.max(), P.min()
P = np.around((P-Pmin)/(Pmax-Pmin),decimals=2)
print('after')
print(P)
```

**Expected Output**:<pre>before
[[0.87 0.58]
 [0.28 0.19]
 [0.41 0.12]] 
<br>after
[[1.   0.62]
 [0.21 0.09]
 [0.39 0.  ]]

### 24. 
**Given a 1D array, negate all elements which are `between -4 and 3`, in place.**<pre> (1 line ★☆☆)

```python
R = np.arange(-8,9)
print('before')
print(R,'\n')

R = np.negative(R,where=np.logical_and(R>=-4,R<=3))
print('after')
print(R)
```

### 25. 
**Get common values between two given arrays**<pre> (1 line ★☆☆) 

```python
np.random.seed(3)
S1 = np.random.randint(0,10,10)
print('S1=', S1,'\n')

np.random.seed(5)
S2 = np.random.randint(0,10,10)
print('S2=', S2,'\n')

S = np.intersect1d(S1,S2)
print('S =', S)
```

<br>S2= [3 6 6 0 9 8 4 7 0 0] 
<br>S = [0 3 8 9]

**Create random uniform vector of `size 12` then replace the maximum value by `1`, and replace the minimum value by `0`**<pre>( 3 lines ★★☆)
